# 04_00A Run Classification Notebooks
Chay 04_01 -> 04_06.

```python
from pathlib import Path
import subprocess, sys
import pandas as pd
PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
NOTEBOOK_DIR = PROJECT_ROOT / 'notebooks'
STOP_ON_ERROR = True
run_list = [
 '04_01_Classification_LogisticRegression.ipynb',
 '04_02_Classification_RandomForest.ipynb',
 '04_03_Classification_NaiveBayes.ipynb',
 '04_04_Classification_LinearSVC.ipynb',
 '04_05_Classification_GBTClassifier.ipynb',
 '04_06_Classification_DecisionTree.ipynb',
]
results = []
for nb in run_list:
    src = NOTEBOOK_DIR / nb
    cmd = [sys.executable, '-m', 'jupyter', 'nbconvert', '--to', 'notebook', '--execute', '--inplace', '--ExecutePreprocessor.kernel_name=python3', str(src)]
    proc = subprocess.run(cmd, cwd=str(PROJECT_ROOT), capture_output=True, text=True)
    if proc.returncode == 0:
        results.append((nb, 'ok', 'executed'))
        print('[OK]', nb)
    else:
        err = (proc.stderr or proc.stdout or '')[:1500]
        results.append((nb, 'failed', err))
        print('[FAILED]', nb)
        if STOP_ON_ERROR:
            break
summary = pd.DataFrame(results, columns=['notebook','status','message'])
print(summary.to_string(index=False))
out = PROJECT_ROOT / 'reports' / 'run_classification_summary.csv'
out.parent.mkdir(parents=True, exist_ok=True)
summary.to_csv(out, index=False)
print('Saved:', out)
```
