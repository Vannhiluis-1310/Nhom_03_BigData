Notebook dieu phoi chay tuan tu 04_01 -> 04_15.

```python
from pathlib import Path
import subprocess
import sys

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
NOTEBOOK_DIR = PROJECT_ROOT / 'notebooks'
EXEC_DIR = PROJECT_ROOT / 'reports' / 'executed_notebooks'
EXEC_DIR.mkdir(parents=True, exist_ok=True)

STOP_ON_ERROR = True
TIMEOUT_SECONDS = 0  # 0 = no timeout

run_list = [
    '04_01_Classification_LogisticRegression.ipynb',
    '04_02_Classification_RandomForest.ipynb',
    '04_03_Classification_NaiveBayes.ipynb',
    '04_04_Classification_LinearSVC.ipynb',
    '04_05_Classification_GBTClassifier.ipynb',
    '04_06_Regression_LinearRegression.ipynb',
    '04_07_Regression_DecisionTreeRegressor.ipynb',
    '04_08_Regression_RandomForestRegressor.ipynb',
    '04_09_Clustering_KMeans.ipynb',
    '04_10_Clustering_BisectingKMeans.ipynb',
    '04_11_Clustering_GaussianMixture.ipynb',
    '04_12_Recommendation_ALS.ipynb',
    '04_13_PatternMining_FPGrowth.ipynb',
    '04_14_Model_Comparison_Summary.ipynb',
    '04_15_Model_Evaluation_Visualization.ipynb',
]

print('Project root:', PROJECT_ROOT)
print('Notebook dir:', NOTEBOOK_DIR)
print('Executed outputs:', EXEC_DIR)
```

```python
results = []

for nb_name in run_list:
    src_nb = NOTEBOOK_DIR / nb_name
    out_nb = EXEC_DIR / nb_name

    if not src_nb.exists():
        msg = f'[MISSING] {src_nb}'
        print(msg)
        results.append((nb_name, 'missing', msg))
        if STOP_ON_ERROR:
            break
        continue

    cmd = [
        sys.executable,
        '-m',
        'jupyter',
        'nbconvert',
        '--to',
        'notebook',
        '--execute',
        '--inplace',
        '--ExecutePreprocessor.kernel_name=python3',
        str(src_nb),
    ]

    print('\n' + '=' * 100)
    print('RUN:', nb_name)
    print('CMD:', ' '.join(cmd))

    try:
        proc = subprocess.run(cmd, cwd=str(PROJECT_ROOT), capture_output=True, text=True, timeout=None if TIMEOUT_SECONDS == 0 else TIMEOUT_SECONDS)
        if proc.returncode == 0:
            results.append((nb_name, 'ok', 'executed'))
            print(f'[OK] {nb_name}')
        else:
            err = (proc.stderr or proc.stdout or '').strip()
            results.append((nb_name, 'failed', err[:2000]))
            print(f'[FAILED] {nb_name}')
            print(err[:2000])
            if STOP_ON_ERROR:
                break
    except Exception as e:
        results.append((nb_name, 'exception', str(e)))
        print(f'[EXCEPTION] {nb_name}: {e}')
        if STOP_ON_ERROR:
            break
```

```python
import pandas as pd
summary = pd.DataFrame(results, columns=['notebook', 'status', 'message'])
print(summary.to_string(index=False))
summary_path = PROJECT_ROOT / 'reports' / 'run_all_models_summary.csv'
summary.to_csv(summary_path, index=False)
print('Saved:', summary_path)
```
